```python
#Introducing Pandas Libraries to the Notebook
import pandas as pd
```

```python
#Loading the Aviation Data File as a Pandas DataFrame
#Assigned to a variable
avdf = pd.read_csv("AviationData.csv", encoding="ISO-8859-1", low_memory=False)
```

```python
#Looking at the basic features of the DataFrame
#Rows,Columns,Data Types,Missing Values
avdf.head()
```

```python
avdf.info()
```

```python
#The data has 88889 rows and 31 columns
avdf.shape
```

```python
#Checking for missing values
avdf.isnull().sum()
```

```python
#Dropping rows with missing values is not recommended since every row will be removed.
avdf.dropna()
```

```python
#Checking for duplicates
avdf.duplicated().sum()
#There are no duplicates in the DataFrame
```

```python
#Dropping rows with missing values in certain columns because they lack essential data
avdf = avdf.dropna(subset=['Publication.Date', 'Report.Status','Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured'])
```

```python
#The essential columns now contain actual values.
avdf.info()
```

```python
#Adjusting the index after dropping some rows
avdf.reset_index(drop = False,inplace=True)
```

```python
avdf['Model'].value_counts()

```

```python
#Finding Aircraft with Minor damage 
avdf.loc[avdf['Aircraft.damage'] == 'Minor']
```

```python
#Finding most safe aircraft
dt3 = avdf.loc[avdf['Aircraft.damage'] == 'Minor']
dt3.loc[dt3['Total.Serious.Injuries']== 0.0]
dt4 = dt3.loc[dt3['Total.Serious.Injuries']== 0.0]
dt4.loc[dt4['Total.Uninjured']== 0.0]
dt5 = dt4.loc[dt4['Total.Uninjured']== 0.0]
dt5.loc[dt5['Total.Minor.Injuries']== 0.0]
dt6 = dt5.loc[dt5['Total.Minor.Injuries']== 0.0]
dt7 = dt6.loc[dt6['Total.Fatal.Injuries']== 0.0]
```

```python
#Viewing models of the 'most safe' aircraft
#most frequent model is MD-11
dt7['Model'].value_counts()
```
